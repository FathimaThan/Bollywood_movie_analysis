# Exploratory Data Analysis

```python
# Visualize the distribution of actor rankings
plt.figure(figsize=(10, 6))
plt.hist(actor['normalizedRating'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Actor Rankings')
plt.xlabel('Normalized Rating')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
top_actors = actor.groupby('actorId')['normalizedRating'].mean().sort_values(ascending = False).head(10)
top_actors
```

```python
top_actors_data = actor[actor['actorId'].isin(top_actors.index)]
top_actors_data
```
